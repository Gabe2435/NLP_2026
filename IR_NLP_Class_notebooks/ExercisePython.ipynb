```python

def third_person(verb_list):
  tp_verbs=[]
  for i in range(len(verb_list)):
    if verb_list[i][-1] in ("s", "ch", "sh", "x", "z", "o"):
      tp_verbs.append(verb_list[i]+"es")
    elif verb_list[i][-1]== "y":
      tp_verbs.append(verb_list[i]+"ies")
    else:
      tp_verbs.append(verb_list[i]+"s")

  return tp_verbs

verbs=["do","sky","walk"]
tp_verbs=[]
tp_verbs=third_person(verbs)
print(tp_verbs)
```

```python
def overlapping_1(l_1,l_2):
  for i in range(len(l_1)):
    for j in range(len(l_2)):
      if l_1[i]==l_2[j]:
        return True
  return False

def overlapping_2(l_1,l_2):
  for i in l_1:
    if i in l_2:
      return True
  return False

l_1=[1,2,3,4,5]
l_2=[6,7,8,9,10]
print(overlapping_1(l_1,l_2))
l_3=[1,2,3,4,5,6]
print(overlapping_1(l_3,l_2))

print(overlapping_2(l_1,l_2))
print(overlapping_2(l_3,l_2))
```

```python
def is_palindrome(string):
  string=string.lower()
  string=string.replace(" ","")
  string=string.replace(",","")
  string=string.replace("'","")
  string=string.replace(".","")
  string=string.replace("!","")
  string=string.replace("?","")

  for i in range(len(string)):
    if string[i]!=string[-i-1]:
      return False
  return True

string_1="Dammit, I'm mad!"
string_2="Was it a rat I saw?"
string_3="Hello World"
print(f"{string_1} \t"+ f"{is_palindrome(string_1)}")
print(f"{string_2} \t"+ f"{is_palindrome(string_2)}")
print(f"{string_3} \t"+ f"{is_palindrome(string_3)}")
```

```python
def is_pangram(string):
  string=string.lower()

  alphabet="abcdefghijklmnopqrstuvwxyz"
  for i in range(len(alphabet)):
    if alphabet[i] not in string:
      return False

  return True

print(is_pangram("The quick brown fox jumps over the lazy dog."))
print(is_pangram("Hello World!"))
```

```python
def char_freq(string):

  frequency_dic={}
  string=string.lower()

  for i in range(len(string)):
    if string[i] in frequency_dic:
      frequency_dic[string[i]]+=1
    else:
      frequency_dic[string[i]]=1
  return frequency_dic

string="Hello World!"
frequency_dic=char_freq(string)
print(frequency_dic)

string="abbacccaaaabbbb"
frequency_dic=char_freq(string)
print(frequency_dic)
```

```python
import random

def anagram(word):
  word_listed = list(word)
  random.shuffle(word_listed)
  word = ''.join(word_listed)
  return word

def is_anagram(word_1, word_2):
  listed_word1=list(word_1)
  listed_word2=list(word_2)

  if len(listed_word1)!=len(listed_word2):
    return False

  for letter in listed_word1:
    if letter in listed_word2:
      listed_word2.remove(letter)
    else:
      return False

  return True

Vocabulary=["dog","sun","moon","guitar","piano"]
random_index = random.randint(0, len(Vocabulary) - 1)

permuted_word=anagram(Vocabulary[random_index])
guess_word=""
while guess_word!=Vocabulary[random_index] and guess_word!="q":
  print("\nGuess the original word!" +"\t type q to stop" + f"\nshuffled word: {permuted_word}")
  guess_word=input("Guess the word: ")

print(is_anagram(guess_word,permuted_word))

print(is_anagram("abcd",permuted_word))
```

```python
#lingo game code
def is_in(guess,word):

  count=[0]*len(word)
  listed_word=list(word)

  for i in range(len(listed_word)):
    if guess[i] in listed_word:
      count[i]+=1
      listed_word.remove(guess[i])

  return count

def correct_position(guess,word):

  count=[0]*len(word)

  for i in range(len(word)):
    if guess[i] == word[i]:
      count[i]+=1

  return count

word="garden"
guess=""
while guess != word:

  guess=input("Guess the word: ")
  count1=is_in(guess,word)
  count2=correct_position(guess,word)
  listed_guess=list(guess)

  for i in range(len(word)):
    if count1[i]>0 and count2[i]>0:
      listed_guess[i]="["+listed_guess[i]+"]"
      p_guess="".join(listed_guess)
    elif count1[i]>0:
      listed_guess[i]="("+listed_guess[i]+")"
      p_guess="".join(listed_guess)
  print(p_guess)

```
