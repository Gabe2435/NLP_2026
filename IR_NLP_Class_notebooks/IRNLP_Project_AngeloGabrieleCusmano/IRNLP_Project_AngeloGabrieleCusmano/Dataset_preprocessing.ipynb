```python
import math
import string
import numpy as np
import pandas as pd

def get_from_file(file_df):
    df = pd.read_csv(file_df)

    # Lista degli speaker di interesse
    selected_speakers = ["Chandler"]

    # Filtro il dataframe
    # Trova tutti i Dialogue_ID che contengono almeno uno speaker selezionato
    dialogue_ids = df[df['Speaker'].isin(selected_speakers)]['Dialogue_ID']

    # Filtra tutto il dataframe per quei Dialogue_ID
    df_filtered = df[df['Dialogue_ID'].isin(dialogue_ids)]
    # Estrazione delle colonne Utterance e Speaker
    x = df_filtered['Utterance'].tolist()
    y = df_filtered['Dialogue_ID'].tolist()
    z = df_filtered['Speaker'].tolist()
    w = df_filtered['Utterance_ID'].tolist()

    return x, y, z, w

def create_LLMdataset(x,y,z,output_filename="LLM_test.csv"):
  prompt=[]
  answer=[]


  for i in range(len(x)):
    if z[i]!="Chandler" and (i+1)<len(x) and z[i+1]=="Chandler" :
      if y[i]==y[i+1]:
        prompt.append(x[i])
        answer.append(x[i+1])
        i+=1

  print(prompt)
  print(answer)

  # Crea un DataFrame con le coppie prompt-answer
  df_out = pd.DataFrame({
       'prompt': prompt,
       'answer': answer
  })

  # Salva il DataFrame in formato CSV
  df_out.to_csv(output_filename, index=False)

  print(f"Dataset salvato come '{output_filename}' con {len(df_out)} esempi.")

  return df_out

```

```python
file_df = '/content/drive/MyDrive/MELD/test_sent_emo.csv'
utt, did, sp, uttid = get_from_file(file_df)

print(utt)
print(did)
```
