```python
#more efficient


#Given two words the code retrieve documents for four queries: word1 AND word2, word1 OR word2, |word1_position-word2_position|<distance and NOT(word1 AND word2)


def query_and(word1,word2):

  docs=[]
  i,j=0,0

  if word1 not in invidx or word2 not in invidx:  #if word1 and word2 not in the vocabulary then it returns None
    return None
  elif word1==word2:                     #if word1=word2 the list of retrieved document for AND query is the inverted index of word1
    for i in invidx[word1]:
      docs.append(i[0])
    return docs

  while i<len(invidx[word1]) and j<len(invidx[word2]):    #more efficient way to avoid using nested for.
    if invidx[word1][i][0]==invidx[word2][j][0]:
      docs.append(invidx[word1][i][0])                  # If the document IDs are the same, add the ID to the result list
      i+=1
      j+=1
    elif invidx[word1][i][0]<invidx[word2][j][0]:
      i+=1                                              #if the document ID in is less than the one in word2, move index i to find a match
    else:
      j+=1

  return docs

def query_or(word1,word2):

  docs=[]
  i,j = 0,0

  if word1 not in invidx and word2 not in invidx:
    return None                                 # If neither word is in the index, return None
  elif word1 not in invidx:
    for i in invidx[word2]:
      docs.append(i[0])      # if word1 isn't in the vocabulary, add the document IDs for word2 to the docs list
    return docs
  elif word2 not in invidx:
    for i in invidx[word1]:
      docs.append(i[0])
    return docs
  elif word1==word2:          #if word1=word2 then the the list of retrieved document for OR query is the inverted index of word1
    for i in invidx[word1]:
      docs.append(i[0])       #for exclusive-or return None
    return docs

  while i<len(invidx[word1]) and j<len(invidx[word2]):
    if invidx[word1][i][0]>invidx[word2][j][0]:
      docs.append(invidx[word2][j][0])
      j+=1
    elif invidx[word1][i][0]<invidx[word2][j][0]:
      docs.append(invidx[word1][i][0])
      i+=1
    else:
      docs.append(invidx[word1][i][0])      # for exclusive-or comment this line
      i+=1
      j+=1

  #adding all the remaining docs

  if i>j:
    while j<len(invidx[word2]):
      docs.append(invidx[word2][j][0])
      j+=1
  else:
    while i<len(invidx[word1]):
      docs.append(invidx[word1][i][0])
      i+=1

  return docs

#get all the indexes of the same word in the same document in order to handle the computation of distances between word1 and word2
def get_indexes(word,i):
  indexes=[]
  for j in range(len(sentences_clean[i])):
    if sentences_clean[i][j]==word:
      indexes.append(j)
  return indexes

#compute all the distance between words even if they appear more than once using obtained indexes
def is_less_than_distance(word1,word2,distance,i):

  indexes_word1=get_indexes(word1,i)
  indexes_word2=get_indexes(word2,i)

  distances=[]
  max_distance=0
  for k in indexes_word1:
    for n in indexes_word2:
      distances.append(abs(k-n))
      if min(distances)<distance:
        return True                     #if it find a distance that respect the condition it return True stopping the iterations

  return None


def query_andnorm(word1,word2,distance):

  docs=[]
  i,j=0,0

  if word1 not in invidx or word2 not in invidx:
    return None
  elif word1==word2:
    for i in invidx[word1]:         #assuming that the distance between a word and itself is 0 and not handling the case in which the same word appears more than once in a document
      docs.append(i[0])
    return docs

  while i<len(invidx[word1]) and j<len(invidx[word2]):
    if invidx[word1][i][0]==invidx[word2][j][0]:
      if invidx[word1][i][1]>1 or invidx[word2][j][1]>1:        # control if a word occurs more than once in the same document
        if is_less_than_distance(word1,word2,distance,invidx[word1][i][0]):  #if one of the possible distance respect the condition then it adds the DOC ID
          docs.append(invidx[word1][i][0])
      elif abs(sentences_clean[invidx[word1][i][0]].index(word1)-sentences_clean[invidx[word1][i][0]].index(word2))<distance:       #doesn't handle cases in which word1 or word2 appear 2 or more times \\solved
        docs.append(invidx[word1][i][0])                                                                                            #2..index() module take the index of the first time the element appear in the list, so it doesn't consider if there is a copy of the word in another position
        #print(f"distance={abs(sentences_clean[invidx[word1][i][0]].index(word1)-sentences_clean[invidx[word1][i][0]].index(word2))}") #evident problem with distance<4 in sentence 16 with words "discrimination" and "equal", the second time equal appears is a distance 1 but is not considered because the first time is at distance 4
      i+=1
      j+=1
    elif invidx[word1][i][0]<invidx[word2][j][0]:
      i+=1
    else:
      j+=1

  return docs

def NOT_word(word):
  not_word=[]
  i=0
  for k in range(len(invidx[word])):
    while i < invidx[word][k][0]:
      not_word.append(i)
      i+=1
    else:
      i+=1
  while i<len(sentences_clean):
      not_word.append(i)
      i+=1

  return not_word

def query_NOTandNOT(word1,word2):

  docs=[]

  if word1 not in invidx and word2 not in invidx:
    return [i for i in range(len(sentences_clean))]
  if word1 not in invidx:
    not_word1=[i for i in range(len(sentences_clean))]
  else:
    not_word1=NOT_word(word1)
  if word2 not in invidx:
    not_word2=[i for i in range(len(sentences_clean))]
  else:
    not_word2=NOT_word(word2)
  if word1==word2:
    docs=not_word1

  i,j=0,0

  while i<len(not_word1) and j<len(not_word2):      #and query for not_word1 and not_word2
    if not_word1[i]==not_word2[j]:
      docs.append(not_word1[i])
      i+=1
      j+=1
    elif not_word1[i]<not_word2[j]:
      i+=1
    else:
      j+=1


  return docs

#input
word1=input("word1: ").lower()
word2=input("word2: ").lower()
distance=int(input(f"distance for query ||{word1}-{word2}||<distance: "))

#word1 = 'discrimination'
#word2 = 'equal'
#distance=10

if word1 in invidx:
  print(f"inverted index {word1}: {invidx[word1]}")
else:
  print(f"inverted index {word1}: None")

if word2 in invidx:
  print(f"inverted index {word2}: {invidx[word2]}")
else:
  print(f"inverted index {word2}: None")

#query_and = word1 and word2

retrieved_docs_and=query_and(word1,word2)
retrieved_docs_or=query_or(word1,word2)
retrieved_docs_andnorm=query_andnorm(word1,word2,distance)
retrieved_docs_notandnot=query_NOTandNOT(word1,word2)

print(f"query: {word1} and {word2} : doc IDs {retrieved_docs_and}")
print(f"query: {word1} or {word2} : doc IDs {retrieved_docs_or}")
print(f"query: {word1} and {word2} in a distance of less than {distance} words: doc IDs {retrieved_docs_andnorm}")
print(f"query: {word1} NOTandNOT {word2} : doc IDs {retrieved_docs_notandnot}")

#alternative:given the two words creates two list associated using the dictionary
```
