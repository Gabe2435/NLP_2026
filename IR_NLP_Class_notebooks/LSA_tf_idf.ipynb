```python
#sentences=text.splitlines()
sentences = [nltk.word_tokenize(x) for x in sentences]

print(sentences)
```

```python
#removing stopwords and punctuation from all the documents in order to avoid useless iteration for word not in the vocabulary

sentences_clean=[]
for i in range(len(sentences)):
  sentences_clean.append([x for x in sentences[i] if x not in stop and x not in string.punctuation])
print(sentences_clean)
```

```python
from itertools import count
import time

invidx={}

#Initializing the vocabulary

for x in words_clean:
  if x not in invidx:
    invidx[x]=[]

#creating the inverted index

for i in range(len(sentences)):
  for x in invidx:
    count=0

    if x in sentences[i]:
      count=sentences[i].count(x)
      invidx[x].append([i,count])   #Number of occurrences is saved too, useful for query: |word1_position-word2_position|<distance, tf-idf
      print(x,i)


print(invidx)

#comments
# the dictionary associate to each key (word) a list of lists that contain DOC ID and the number of occurrences in that document
print(invidx["whereas"][1][0]) #print the doc ID in position 1 of the list associated to "whereas"
print(invidx["whereas"][1][1]) #print the number of occurences of the word in the relative document
```

```python
##### TF #####

#for the tf part I already have the inverted index that stores for each word the doc ID and frequency
#just use formula tf_ij=1+log(1+f_ij)

##### IDF #####

#for this I can just use the invidx too. I compute the total number of documents in which a word occur from invidx
#Once all the TF-IDF terms have been calculated, they can be stored in a text file to be retrieved later for queries.

import math

#I decided to use a dictionary; an alternative could have been using a matrix or a list.
#In the dictionary, I used the words as keys, but the respective index of each word could have been used instead.

tf_idf={}

for word in invidx:
  tf_idf[word]=[]
  idf=math.log2(len(sentences_clean)/len(invidx[word]))
  for doc_ID,f_ij in invidx[word]:
    tf=1+math.log2(1+f_ij)
    tf_idf[word].append([doc_ID,tf*idf])

print(tf_idf)

```

```python
#Defining documents vectors

doc_vector=[{} for _ in range(len(sentences_clean))]
norms=[0]*len(doc_vector)

for word in tf_idf:
  for i,w in tf_idf[word]:
    doc_vector[i][word]=w       #each component of doc_vector is a specific document vector. Each document vector is a dictionary that associate the weight to a word
    norms[i]+=w**2

print(f"Documents vectors before normalization : {doc_vector}")

for i in range(len(doc_vector)):
  if norms[i]>0:
    norm = math.sqrt(norms[i])
    for word in doc_vector[i]:
      doc_vector[i][word]/=norm

print(f"Documents vectors after normalization : {doc_vector}")

#given a query, compute the distances between the query and the documents.

def query_distance(query,doc_vector):
  similarities=[]

  for i in range(len(doc_vector)):
    similarity=0
    for word in query:
      if word in doc_vector[i]:
        similarity+=query[word]*doc_vector[i][word]
    if similarity!=0:
      similarities.append((similarity,i))

  similarities.sort(reverse=True)
  return similarities

#query=input()
query="rights and freedom"
query_vector={}

#processing the query removing stopwords and punctuation

query_words=[x for x in nltk.word_tokenize(query.lower()) if x not in stop and x not in string.punctuation]

for word in query_words:
  if word in tf_idf:
    query_vector[word]=(1+math.log2(1+query_words.count(word)))*math.log2(len(sentences_clean)/len(invidx[word]))

query_norm = math.sqrt(sum(w ** 2 for w in query_vector.values()))

if query_norm > 0:
    for word in query_vector:
        query_vector[word] /= query_norm

print(query_vector)

similarities = query_distance(query_vector, doc_vector) if query_vector else "Query vector is empty, no similarity computed"

for sim,doc_ID in similarities:
  print(f"Similarity with document {doc_ID}: {round(sim,5)}")
```

```python
##### LSA #####

#not using sparse matrix
####! Number of Occurencies used !####

#####    NEXT CELL USES TF_IDF   #####

import numpy as np
import pandas as pd

TD=np.zeros((len(invidx),len(sentences_clean)))  #term-document matrix

word_index = {word: i for i, word in enumerate(invidx)}

for i,word in enumerate(invidx):    #enumerate gives a tuple with index and word
  for index,value in invidx[word]:
    TD[i,index]=value

U,S,VT=np.linalg.svd(TD,full_matrices=False)

query="freedom"

query_words=[x for x in nltk.word_tokenize(query.lower()) if x not in stop and x not in string.punctuation]
query_vector=np.zeros(len(tf_idf))

for word in query_words:
  if word in invidx:
    i=word_index[word]
    query_vector[i]=1

for r in [5,10,15,20]:
  print(f"\n-------------\nr={r}")
  Ur=U[:,:r]
  Sr=np.diag(S[:r])
  VTr=VT[:r,:]  #it has document vectors in columns in the semantic space of dimension r

  TDr=Ur@Sr@VTr

  query_svd=query_vector.T@Ur@np.linalg.inv(Sr)

  similarities=[]

  for i in range(len(sentences_clean)):
    similarity=np.dot(query_svd/np.linalg.norm(query_svd),VTr[:,i]/np.linalg.norm(VTr[:,i]))
    similarities.append((similarity,i))

  similarities.sort(reverse=True)

  i=0
  while i<10:
    print(f"\nSimilarity with document {similarities[i][1]}: {round(similarities[i][0],5)}")
    print(f"Document {similarities[i][1]}: {text.splitlines()[similarities[i][1]]}")
    i+=1

    ### CORRELATION AND DATA FRAME ###

  # Calculate the correlation matrix
  correlation_matrix = np.corrcoef(TD)
  columns_norms = np.linalg.norm(TDr, axis=1, keepdims=True)
  TDr_normalized = TDr / columns_norms
  correlation_matrix = TDr_normalized @ TDr_normalized.T
  # Convert to DataFrame for visualization and association of each row and column with the respective word
  correlation_df = pd.DataFrame(correlation_matrix, index=list(invidx.keys()), columns=list(invidx.keys()))
  term = "freedom"    #check correlation with only this term

  corr_series = correlation_df.loc[term]

  corr_series_sorted = corr_series.sort_values(ascending=False)

  # Show first 10 terms with highest correlation
  print(f"\n10 terms with highest correlation\n")
  print(corr_series_sorted.head(10))

  terms = ["equality", "equal", "freedom","free","freely","rights"]  #check correlation between these terms
  df_subset = correlation_df.loc[terms, terms]
  #subset with the specified terms
  print("\nSubset Correlation\n")
  print(df_subset)
```
